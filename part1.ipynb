```
#Q3 Write a function that creates an incremental array of dimension (1,n) with values
#between 0 and 1. Use arr.shape to verify.
#ar3 = np.linspace(0, 1, 7) --- print(c, c.shape)
def fun_n(n):
    tmp = np.linspace(0, 1, n).reshape(1, n)
    return tmp
ar3 = fun_n(5)
print(ar3, ar3.shape)
```

```
#Q5 Using the function in Q3, get m vectors and bind them together (to have a m x n)
#matrix. Plot the matrix with matplotlibâ€™s imshow
def fun_m(m, n):
    arr = np.array(fun_n(n))
    for num in range(m-1):
        tmp = np.array(fun_n(n))
        newarray = np.append(arr, tmp, axis = 0 )
        arr = newarray
    print(newarray, newarray.shape)
    return newarray

mat1 = fun_m(3, 4)
plt.imshow(mat1, interpolation='bilinear')
```
